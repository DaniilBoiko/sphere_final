### Docs & titles

```python
import gzip
from tqdm.notebook import tqdm
```

```python
mapper = {}

with gzip.open('data/docs.lemm.tsv.gz', 'rt') as f:
    for line in tqdm(f):
        splitted = line.split('\t')
        mapper[int(splitted[0])] = line.split('\t')[3].lower()
```

```python
import sys
sys.getsizeof(mapper)
```

```python
with open('data/title_mapper.pkl', 'wb') as f:
    pkl.dump(mapper, f)
```
