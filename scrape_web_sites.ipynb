```python
import pandas as pd
import requests
import bs4
import eventlet
import pickle as pkl
```

```python
url_data = pd.read_csv('data/url.data/url.data', sep='\t', header=None)
url_data = url_data[1].to_dict()
```

```python
def scrape(url):
    links = []
    
    try:
        with eventlet.Timeout(10):
            res = requests.get("http://" + url, timeout=5)
            return res.text
    except:
        return None
    
    return None
```

```python
def save(url, name):
    data = scrape(url)
    
    if data is not None:
        with open(f'data/docs/{name}.pkl', 'wb') as f:
            pkl.dump(data, f)
```

```python
from joblib import Parallel, delayed

Parallel(n_jobs=8)(delayed(save)(url, name) for name, url in url_data.items())
```
